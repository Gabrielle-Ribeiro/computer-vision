# Processamento de Imagens

## 1) O que é uma imagem?

- Uma imagem é uma matriz de pixels.
    - **Pixel:** É a menor unidade que representa uma cor em um display.           

- Uma imagem pode ser considerada uma função. Essa função transforma pixels em cores.            

- Uma imagem pode ser definida como uma amostragem de um sinal luminoso em um determinado momento, captado por um sensor (a câmera).

## 2) Sintetizando imagens

Com o uso da biblioteca pillow, criaremos uma imagem sintética.

```python
def triangle(size):
    WHITE = (255, 255, 255)
    BLACK = (0, 0, 0)
    
    image = Image.new("RGB", (size, size), WHITE)
    
    for x in range(size):
        for y in range(size):
            if x < y:
                image.putpixel((x, y), BLACK)
    return image
```

```python
triangle_image = triangle(700)
display(triangle_image)
```

```python
def bandeira_franca(height):
    BLUE = (0, 85, 164)
    WHITE = (255, 255, 255)
    RED = (239, 65, 53)
    
    width = 3 * height // 2
    
    image = Image.new("RGB", (width, height), WHITE)
    
    offset = width//3
    
    for x in range(offset):
        for y in range(height):
            image.putpixel((x, y), BLUE)
            image.putpixel((x + 2*offset, y), RED)
            
    return image
```

```python
bandeira_img = bandeira_franca(700)
display(bandeira_img)
```

```python
def bandeira_japao(height):
    width = 3 * height // 2
    
    RED = (173, 35, 51)
    WHITE = (255, 255, 255)
    
    r = 3*height//10
    c = (width//2, height//2)
    
    image = Image.new("RGB", (width, height), WHITE)
    
    for x in range(c[0]-r, c[0]+r):
        for y in range(c[1]-r, c[1]+r):
            if (x-c[0])**2 + (y-c[1])**2 <= r**2:
                image.putpixel((x, y), RED)
                
    return image
```

```python
japao_img = bandeira_japao(700)
display(japao_img)
```
